## Data Description

The dataset used for this analysis consists of a comprehensive list of all movies and TV shows available on Netflix as of mid-2021.  It has 8,800+ rows and 12 columns that provide metadata for each title:

* show_id: A unique identifier for every movie or TV show in the dataset.
* type: Categorizes the content as either a "Movie" or a "TV Show".

* title: The official name of the movie or television series.

* director: The individual(s) responsible for directing the content.

* cast: The lead actors and actresses featured in the production.

* country: The country or countries where the content was produced.

* date_added: The specific date when the title was first made available on the Netflix platform.

* release_year: The actual year the movie or TV show was originally released to the public.

* rating: The age-based content rating (e.g., TV-MA, PG-13, TV-14).

* duration: The length of the content, measured in minutes for movies or number of seasons for TV shows.

* listed_in: The genres or categories assigned to the title (e.g., Documentaries, International TV Shows, Comedies).

* description: A brief summary or synopsis of the plot.

## 3. Data Cleaning

```python
#CLeaning the data.
#Director column
df["director"] = df["director"].fillna("Unknown")
# cast column
df["cast"] = df["cast"].fillna("Unknown")
```

```python
df["country"] = df["country"].fillna("Unknown")

df["country_split"] = df["country"].apply(lambda x: x.split (',')[0]) 
df.info()
#.apply() tells pandas to go through every single row in the country column.
#x x in this case is the variable that's holding our original, messy country column. 
#lambda is a shortcut for creating a tiny, one-time use function. 
#Instead of defining a permanent function use lambda.allowed us to quickly grab only the primary country
```

```python
df["country_split"].value_counts()
```

# Observations
* Replaced the missing country values with unkwown.
* Initially, our 'Country' data was quite cluttered.
Many titles were listed under multiple nations (e.g., 'USA, India, UK'), which created hundreds of unique, overlapping categories that were difficult to graph. To fix this, I simplified the data by extracting only the Primary Country (the first one listed) for each title. This transformed our messy list into a clean, actionable map of Netflix’s main production hubs.

To do this you must first fill the unknown values then use lambda to avoid your code crushing.

```python
# Cleaning the date column
# In info function it was running date as object yet it's supposed to be date.

df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
# errors = "coerce" this tells python that if it finds a value that can't be converted to date don't crash it into NaT. 

df = df.sort_values('release_year')
df["date_added"] = df["date_added"].ffill()
df.info()
```

**Explanation:** 

* We are changing object to date because you cannot do math with letters. Changing it to a Date format allows the computer to understand the timeline, sort the months correctly, and create growth charts.
* I sorted the list by the year the movie was made. This helped me fill in the missing dates more accurately. By looking at movies from the same time, I avoided making date mistakes (like giving an old movie a very new date).
* We are using fill and not drop to  keep your Total Count perfect. If you "Drop," those 10 movies disappear from your project forever. By using "Fill," you keep every single movie in your dataset while giving the missing ones a logical "guess" for their date.

```python
#Cleaning the Rating column
#str.contains('min') checks whether if in the rating there contains the letters "min"'
#na = false means if there are empty cells in the rating column say false don't crash the code
mask = df['rating'].str.contains('min', na=False)

#df.loc = location. This allows us to access specific rows and columns at the same time.
#It tells pandas to find only the rows where mask is True and look specifically at the duration column.
#if true then take the values from the rating column and copy them into the duration column.
df.loc[mask, 'duration'] = df.loc[mask, 'rating']
```

```python
df["rating"] = df["rating"].fillna("Unknown")
df.info()
```

### Explanation of the above process
The Rating column required a two-phase cleanup. 
* First, I performed a conditional migration to move duration values(minutes) from the rating column back to their correct field, ensuring no data loss.
* Second, I filled the remaining nun values with unknown. This ensures that the data is honest. This acknolwdges that many international or independent titles may not have a standard US rating, rather than "guessing" and potentially biasing the distribution. 

## 4. Data Visualization 

## Box Plot To show outliers in our dataset.
I have chosen to use the Movies subset to detect outliers in the release_year for two primary reasons:
   * Movies make up almost 70% of everything on Netflix. Because of this it is easier to spot the outliers that stand out from the rest.
   * A movie is usually released in one specific year. They give us a much cleaner timeline to see how far back the Netflix library really goes. TV-Shows are trickier since they span decades across multiple seasons.

1. Release year:

```python
movies_df = df[df["type"] == "Movie"].copy() #.copy(),tells Pandas to create a brand-new,independent object in your computer's memory.
# Create a Box Plot for Release Year
oldest_year = movies_df['release_year'].min()
print(f"The oldest movie year is: {oldest_year}")
newest_year = movies_df['release_year'].max()
print(f"The newest movie year is: {newest_year}")
print(f"The actual median is: {movies_df['release_year'].median()}")
print("****" * 25)

plt.figure(figsize=(12, 6))

sns.boxplot(x=movies_df['release_year'], color='lightcoral')

# 1. Finding the min and max years
min_year = int(movies_df['release_year'].min())
max_year = int(movies_df['release_year'].max())

# 2. Round the start year down to the nearest decade (e.g., 1942 -> 1940)
start_tick = (min_year // 10) * 10
# 3. Round the end year up to the nearest decade (e.g., 2024 -> 2030)
end_tick = ((max_year // 10) + 1) * 10

# 4. Set the ticks at 10-year intervals
plt.xticks(np.arange(start_tick, end_tick + 1, 10))
plt.title('Identifying Outliers in Release Year (10-Year Intervals)')
plt.xlabel('Release Year')
plt.grid(axis='x', linestyle='--', alpha=0.6) # Optional: adds lines to help see the year

# Use savefig instead of show in our environment
plt.savefig('boxplot_release_year_10yr.png')
```

## Observation from the above box plot.
* The x-axis was scaled from 1940-2030 to maintain a consistent 10-year interval and provide visual clarity for these early outliers.
* Movies released before 2002 (where the left whisker ends) is considered an outlier.
* The box (which contains 50% of netflix movies ) is situated entirely between 2012 and 2018. This tells us that the majority of the movies in this dataset are very modern.
* The line inside the box sits around 2016 indicating that half of all movies in Netflix were released between 2016 and 2021.
* The long tail of "dots" extending to the left indicates a skewed distribution. Even though there's a  huge volume of new movies, Netflix has a very thin, sparse collection of historical firms.
* The box-plot reveals that the movie collection is heavily concentrated in the last decade, with a median release year of 2016. Films released before 2002 are identified as outliers, including a rare selection of historical titles dating back to 1942. 

### Box plot to visualize outliers in duration. 

```python
# cleaning
#1. Extracting number from duration string
movies_df["duration_min"] = movies_df["duration"].str.replace("min", " ")
#2. Converts to numerical values(errors="coerce" handles any weird empty values
movies_df["duration_min"] = pd.to_numeric(movies_df["duration_min"],errors = "coerce")
#3. droping rows where rows might be missing. This is optinal but highly recommended
movies_df = movies_df.dropna(subset=["duration_min"])

# visualizing
plt.figure(figsize=(12,6))
sns.boxplot(x=movies_df["duration_min"], color='gold')

plt.title("Identifying outliers in Movie durations")
plt.xlabel('Duration (minutes)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.show()
```

#### Observations.
* The majority (50%) of movies fall within a tight window of 85 to 110 minutes, suggesting a platform reference for standard feature length content that fits a typical viewing window.
* The line inside the box sits at around 95 minutes. This is the median duration, indicating that half of the movie library consists of films under 95 minutes (1 hour and 35 minutes).
* The distribution is right-skewed,(positive skew). While the volume of movies begins to drop off sharply at the 120-minute mark (aligning with standard industry completion-rate goals), the "tail"  of dots extending towards 200+ minutes shows that "Epic" films exist but are statistically rare.
* There are also outliers on the left side which represent short films or documentaries that are much shorter than a standard feature.

## 3. Proportion of Movies vs. TV Shows

## 5. Content Growth Over Time.
Visualizing growth over time turns a simple list of numbers into a clear timeline of Netflix's history. Instead of just seeing how much content exists today, this chart shows exactly when the library expanded, allowing us to identify the major "pivot point" around 2016 when Netflix transformed from a small library into a global content factory.

```python
#Group by year and type, then count the entries
# unstack() moves 'Movie' and 'TV Show' into separate columns
#.fillna(0): This ensures that if a year had zero TV shows added, the code doesn't crash due to missing data.
growth_df = df.groupby(['year_added', 'type']).size().unstack().fillna(0).fillna(0)
plt.figure(figsize=(14, 7))

# Plot Movies
sns.lineplot(data=growth_df, 
             x=growth_df.index, 
             y='Movie', 
             color='#E50914', 
             label='Movies', 
             linewidth=3, 
             marker='o')
# Plot TV Shows
sns.lineplot(data=growth_df, 
             x=growth_df.index, 
             y='TV Show', 
             color='#221F1F', 
             label='TV Shows', 
             linewidth=3, 
            marker='s')

plt.title('The Growing Gap: Movies vs. TV Shows', fontsize=16, fontweight='bold')
plt.ylabel('Titles Added')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## Observations
* From this line chart, the red line( Movies) is consistently much higher than the black line(TV Shows). This confirms that Movies make up the majority of the content. Even though both are growing, the gap between them remains large, supporting our 2:1 ratio.
* The chart is almost a flat one near zero from (2008 to 2014) because during these years, Netflix was primarily a DVD-by-mail service. The "Streaming" library as tiny and mostly consisted of old movies licensed from other studios. There was very little "fresh" content being added.
* You can see both lines start to climb sharply around 2016. This represents Netflix’s Global Expansion, where they launched in 130 countries at once. To keep all those new international subscribers happy, they had to "flood" the platform with thousands of new titles 
* The "TV Show" Acceleration
Looking at the Black Line (TV Shows) in the later years, even though it is lower than the movie line, its steepness shows it is growing steadily. Netflix realized that a movie keeps a person busy for 2 hours, but a TV show keeps them on the app for 10+ hours. This is their strategy to keep people paying for subscriptions every month.
* The 2020-2021 "Saturation" or "Production Dip"
The line start to flatten out after 2019 due to the impact of the COVID-19 pandemic. Film sets worldwide were shut down. While everyone wanted to watch Netflix, film sets were shut down. Netflix couldn't "supply" as much new content because they couldn't physically film new movies and shows.
* The line chart reveals that Netflix's content strategy is divided into two distinct eras: the Distribution Era (pre-2015), characterized by stagnant growth, and the Production Era (2016-present), marked by an exponential grwoth in titles. The steady rise of the TV Show line shows their goal to increase "watch time" through episodic series.

# 6. Distribution of Content Ratings.
Visualizing the distribution of ratings allows us to understand the "maturity profile" of the platform and identify who Netflix is building its library for. Instead of just listing categories like TV-MA or PG-13, this visualization shows us at a glance whether the platform leans toward mature audiences or family-friendly content

```python
rating_counts = df['rating'].value_counts()

#Grouped the smaller ratings (less than 1%) into 'Others' to keep the pie clean
threshold = 0.01 * len(df)
major_ratings = rating_counts[rating_counts > threshold]
minor_ratings_sum = rating_counts[rating_counts <= threshold].sum()

if minor_ratings_sum > 0:
    major_ratings['Others'] = minor_ratings_sum

plt.figure(figsize=(10, 6))
colors = sns.color_palette('Reds_r', len(major_ratings))

plt.pie(major_ratings, 
        labels=major_ratings.index, 
        autopct='%1.1f%%', 
        startangle=140, 
        colors=colors, 
        pctdistance=0.85, 
        explode=[0.05 if i == 0 else 0 for i in range(len(major_ratings))]) # Pop out the largest slice

#Added a center circle to make it a Donut Chart.
center_circle = plt.Circle((0,0), 0.70, fc='white')
plt.gcf().gca().add_artist(center_circle)

plt.title('Distribution of Content Ratings on Netflix', fontsize=12, fontweight='bold')
plt.axis('equal') 
plt.tight_layout()
plt.show()
```

# Observations.
* TV-MA (Mature Audience) has the largest slice. Netflix spend billions on "Edgy" content because it creates most buzz on social media and wins awards. Adults are the ones who pay the monthly subscriptions bill, so Netflix prioritizes their tastes first to keep them from cancelling.
* TV-14 (Parents strongly cautioned) is the second largest slice. It's mature enough for adult to enjoy, but safe enough for teenagers. This category drives the most "Binge-watching" because it appeals to the widest possible age group. 
* TV-PG (Parental Guidance) is the third slice. This category prevents churn (people cancelling their subscriptions). Even if an adult finishes their TV-MA drama, they won't cancel the subscriptions since there's enough TV-PG content for the family to watch together.
* Ratings contributing less than 1% of the total library, such as TV-Y and G, were grouped into an 'Others' category to maintain visual clarity. This grouping reveals a significant strategic insight: Netflix has a relatively small footprint in the 'Toddler' and 'General Audience' space.
Netflix provides just enough kids' content to be a convenience, but it is not their primary goal. By consolidating these minor slices, the chart more effectively highlights Netflix's core identity as a platform dominated by Mature (TV-MA) and Teen (TV-14) content.

The rating distribution confirms that Netflix's catalog is heavily skewed toward mature Audience. With TV-MA and TV-14 dominating the library, the platform prioritizes edgy, complex storytelling over traditional family-friendly content.
